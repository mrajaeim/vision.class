```python
import cv2
```

```python
cap = cv2.VideoCapture("rstp://wowzaec2demo.streamlock.net/vod/mp4:BigBuckBunny_115k.mp4")

while 1:
    ret, frame = cap.read()
    if not ret:
        print("Can't read frame")
        break
    cv2.imshow("CCTV", frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break


cv2.destroyAllWindows()
cap.release()
```
