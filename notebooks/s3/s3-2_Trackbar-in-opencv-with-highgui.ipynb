```python
window_name = "Color Picker"
red_tracker_name = "Red"
green_tracker_name = "Green"
blue_tracker_name = "Blue"
max_tracker_val = 255

def donothing(*args):
    pass

cv2.namedWindow(window_name)
cv2.createTrackbar(red_tracker_name, window_name, 0, max_tracker_val, donothing)
cv2.createTrackbar(green_tracker_name, window_name, 0, max_tracker_val, donothing)
cv2.createTrackbar(blue_tracker_name, window_name, 0, max_tracker_val, donothing)

img = np.zeros((512,512,3), np.uint8)

while 1:
    # img[...,0] = cv2.getTrackbarPos(blue_tracker_name, window_name) or 0
    # img[...,1] = cv2.getTrackbarPos(green_tracker_name, window_name) or 0
    # img[...,2] = cv2.getTrackbarPos(red_tracker_name, window_name) or 0
    b = cv2.getTrackbarPos(blue_tracker_name, window_name)
    g = cv2.getTrackbarPos(green_tracker_name, window_name)
    r = cv2.getTrackbarPos(red_tracker_name, window_name)
    img[:] = [b, g, r]
    cv2.imshow(window_name, img)
    
    if cv2.waitKey(1) == 27:
        break

cv2.destroyAllWindows()
plt.imshow(img[..., ::-1])
```
