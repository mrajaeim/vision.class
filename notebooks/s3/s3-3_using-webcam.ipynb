```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture(0)

while 1:
    ret, frame = cap.read()
    cv2.imshow("WebCam", frame)
    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

while 1:
    ret, frame = cap.read()
    fi = cv2.flip(frame, 1)
    cv2.imshow("Fliped", fi)
    cv2.imshow("WebCam", frame)
    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

while 1:
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)
    blank_frame = np.zeros(frame.shape[:2], np.uint8)
    
    b,g,r = cv2.split(frame)
    red_frame = cv2.merge([blank_frame, blank_frame, r])
    green_frame = cv2.merge([blank_frame, g, blank_frame])
    blue_frame = cv2.merge([b, blank_frame, blank_frame])

    cv2.imshow("Red", red_frame)
    cv2.imshow("Green", green_frame)
    cv2.imshow("Blue", blue_frame)
    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
