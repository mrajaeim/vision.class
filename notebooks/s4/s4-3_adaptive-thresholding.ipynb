```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
sudoku_img = cv.imread("../../images/sudoku.jpg", 0)

blured_sudoku_img = cv.GaussianBlur(sudoku_img, (5,5), 0)

plt.figure(figsize=(20,16))
plt.subplot(3,4,1); plt.imshow(sudoku_img, cmap="gray", vmin=0, vmax=255); plt.title("Sudoku Original")
plt.subplot(3,4,2); plt.imshow(blured_sudoku_img, cmap="gray", vmin=0, vmax=255); plt.title("Blured Sudoku")

ter, res = cv.threshold(sudoku_img, 127, 255, cv.THRESH_BINARY)
plt.subplot(3,4,5); plt.imshow(res, cmap="gray", vmin=0, vmax=255); plt.title("Sudoku Binary")

ter, res = cv.threshold(sudoku_img, 127, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
plt.subplot(3,4,6); plt.imshow(res, cmap="gray", vmin=0, vmax=255); plt.title("Sudoku Otsu")

ter, res = cv.threshold(blured_sudoku_img, 127, 255, cv.THRESH_BINARY)
plt.subplot(3,4,7); plt.imshow(res, cmap="gray", vmin=0, vmax=255); plt.title("Blured Sudoku Binary")

ter, res = cv.threshold(blured_sudoku_img, 127, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
plt.subplot(3,4,8); plt.imshow(res, cmap="gray", vmin=0, vmax=255); plt.title("Blured Sudoku Otsu")

# Adaptive

res = cv.adaptiveThreshold(sudoku_img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
plt.subplot(3,4,9); plt.imshow(res, cmap="gray", vmin=0, vmax=255); plt.title("AT GausBinary Sudoku")

res = cv.adaptiveThreshold(sudoku_img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
plt.subplot(3,4,10); plt.imshow(res, cmap="gray", vmin=0, vmax=255); plt.title("AT MeanBinary Sudoku")

res = cv.adaptiveThreshold(blured_sudoku_img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
plt.subplot(3,4,11); plt.imshow(res, cmap="gray", vmin=0, vmax=255); plt.title("AT GausBinary Blured Sudoku")

res = cv.adaptiveThreshold(blured_sudoku_img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
plt.subplot(3,4,12); plt.imshow(res, cmap="gray", vmin=0, vmax=255); plt.title("AT MeanBinary Blured Sudoku")

plt.tight_layout()
```
