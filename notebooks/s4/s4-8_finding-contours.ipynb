```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("../../../class.vision/images/shapes.png")
grayscale_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
_, binary_img = cv.threshold(grayscale_img, 127, 255, cv.THRESH_BINARY)

plt.figure(figsize=(10,10))
plt.subplot(121); plt.imshow(grayscale_img, cmap='gray', vmin=0, vmax=255)
plt.subplot(122); plt.imshow(binary_img, cmap='gray', vmin=0, vmax=255)
```

```python
contours, hierarchy = cv.findContours(binary_img, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
temp_img = img.copy()
cv.polylines(temp_img, [contours[3]], True, (0, 0, 255), 2)
plt.imshow(temp_img[...,::-1], vmin=0, vmax=255)
```

```python
cv.polylines(temp_img, [contours[1]], True, (0, 255, 0), 2)
plt.imshow(temp_img[...,::-1], vmin=0, vmax=255)
```

```python
contours, hierarchy = cv.findContours(binary_img, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
temp_img = img.copy()
cv.polylines(temp_img, contours, True, (0, 255, 0), 2)
plt.imshow(temp_img[...,::-1], vmin=0, vmax=255)
```

```python
contours, hierarchy = cv.findContours(binary_img, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
temp_img = img.copy()
cv.drawContours(temp_img, contours, -1, (255, 255, 0), 2)
plt.imshow(temp_img[...,::-1], vmin=0, vmax=255)
```

```python
contours, hierarchy = cv.findContours(binary_img, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
temp_img = img.copy()
cv.drawContours(temp_img, contours, 3, (255, 0, 255), 2)
plt.imshow(temp_img[...,::-1], vmin=0, vmax=255)
```

```python
image = cv.imread('../../../class.vision/images/rectangle.png', 0)
ret, thresh = cv.threshold(image, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)

print("cv.CHAIN_APPROX_NONE:")
print("contours[0].shape: {} so number of points for the contour: {}".format(contours[0].shape, len(contours[0])))

bgr_image = cv.cvtColor(image, cv.COLOR_GRAY2BGR)
for point in contours[0]:
    cv.circle(bgr_image, (point[0][0], point[0][1]), 2, (0,255,0), -1)
plt.imshow(bgr_image[...,::-1]); plt.title('CHAIN_APPROX_NONE')
print('-'*5)


contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
print("cv.CHAIN_APPROX_SIMPLE:")
print("contours[0].shape: {} so number of points for the contour: {}".format(contours[0].shape, len(contours[0])))

bgr_image = cv.cvtColor(image, cv.COLOR_GRAY2BGR)
for point in contours[0]:
    cv.circle(bgr_image, (point[0][0], point[0][1]), 2, (0,255,0), -1)

plt.figure(); plt.imshow(bgr_image[...,::-1]); plt.title('CHAIN_APPROX_SIMPLE')
```
