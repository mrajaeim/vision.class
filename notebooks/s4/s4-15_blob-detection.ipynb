```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
blob_img = cv.imread("../../../class.vision/images/blob.jpg")
blob_size_img = cv.imread("../../../class.vision/images/blob_size.jpg")
blob_circularity_img = cv.imread("../../../class.vision/images/blob_Circularity.jpg")
blob_convexity_img = cv.imread("../../../class.vision/images/blob_convexity.jpg")
blob_inertia_img = cv.imread("../../../class.vision/images/blob_inertia.jpg")
blob_threshold_img = cv.imread("../../../class.vision/images/blob_Threshold.jpg")

plt.figure(figsize=(10,10))
plt.subplot(321); plt.imshow(blob_img[...,::-1])
plt.subplot(322); plt.imshow(blob_size_img[...,::-1])
plt.subplot(323); plt.imshow(blob_circularity_img[...,::-1])
plt.subplot(324); plt.imshow(blob_convexity_img[...,::-1])
plt.subplot(325); plt.imshow(blob_inertia_img[...,::-1])
plt.subplot(326); plt.imshow(blob_threshold_img[...,::-1])
```

```python
params = cv.SimpleBlobDetector_Params()
detector = cv.SimpleBlobDetector_create(params)

keypoints = detector.detect(blob_img)

result1 = cv.drawKeypoints(blob_img, keypoints, np.array([]))
result2 = cv.drawKeypoints(blob_img, keypoints, np.array([]), flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(blob_img[...,::-1]);plt.title("image");
plt.subplot(132);plt.imshow(result1[...,::-1]);plt.title("result1");
plt.subplot(133);plt.imshow(result2[...,::-1]);plt.title("result1");
```

```python
params = cv.SimpleBlobDetector_Params()
params.filterByArea = True
params.minArea = 2000
params.maxArea = 20000
detector = cv.SimpleBlobDetector_create(params)

keypoints = detector.detect(blob_size_img)

result = cv.drawKeypoints(
    blob_size_img,
    keypoints,
    np.array([]),
    color=(0, 0, 255),
    flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
)

plt.imshow(result[...,::-1])
```

```python
params = cv.SimpleBlobDetector_Params()
params.filterByCircularity = True
params.minCircularity = 0.83
detector = cv.SimpleBlobDetector_create(params)

keypoints = detector.detect(blob_circularity_img)

result = cv.drawKeypoints(
    blob_circularity_img,
    keypoints,
    np.array([]),
    color=(0, 0, 255),
    flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
)

plt.imshow(result[...,::-1])
```
