```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
old_ir_coins = cv.imread("../../../class.vision/images/old-iranian-coins.jpg")
money_img = cv.imread("../../../class.vision/images/money.jpg")
ir_coins_img = cv.imread("../../../class.vision/images/iranian-coin.jpg")

plt.figure(figsize=(10,10))
plt.subplot(131); plt.imshow(old_ir_coins[...,::-1])
plt.subplot(132); plt.imshow(money_img[...,::-1])
plt.subplot(133); plt.imshow(ir_coins_img[...,::-1])
```

```python
def is_valid_ellipse(ellipse):
    return ellipse[1][0] > 0 and ellipse[1][1] > 0
```

```python
gray_img = cv.cvtColor(old_ir_coins, cv.COLOR_BGR2GRAY)
binary_img = cv.adaptiveThreshold(gray_img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY_INV, 11, 2)
closing = cv.morphologyEx(binary_img, cv.MORPH_CLOSE, cv.getStructuringElement(cv.MORPH_ELLIPSE, (7, 7)))

contours, _ = cv.findContours(closing, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

draw_img = old_ir_coins.copy()

for cnt in contours:
    if len(cnt) < 5:
        continue
    ellipse = cv.fitEllipse(cnt)
    if is_valid_ellipse(ellipse):
        cv.ellipse(draw_img, ellipse, (0, 255, 0), 2)

plt.figure(figsize=(10,10))
plt.subplot(131); plt.imshow(binary_img, cmap='gray'); plt.title("Binary Image")
plt.subplot(132); plt.imshow(closing, cmap='gray'); plt.title("Closing Image")
plt.subplot(133); plt.imshow(draw_img[...,::-1]); plt.title("Annotated Coins")
```

```python
# Load image
img = ir_coins_img.copy()
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# --- Preprocessing ---
# Remove noise but keep edges
gray = cv.medianBlur(gray, 5)
gray = cv.bilateralFilter(gray, 9, 75, 75)

# Improve contrast
gray = cv.equalizeHist(gray)

# --- Hough Circle Detection ---
circles = cv.HoughCircles(
    gray,
    cv.HOUGH_GRADIENT,
    dp=1.2,             # accumulator resolution
    minDist=90,         # distance between coin centers
    param1=120,         # Canny high threshold
    param2=35,          # accumulator threshold (lower  more sensitive)
    minRadius=40,
    maxRadius=130
)

# --- Draw results ---
output = img.copy()

if circles is not None:
    circles = np.uint16(np.around(circles))
    for (x, y, r) in circles[0]:
        cv.circle(output, (x, y), r, (0, 255, 0), 2)
        cv.circle(output, (x, y), 2, (0, 0, 255), 3)

print(f"Detected coins: {0 if circles is None else len(circles[0])}")

plt.figure(figsize=(10, 6))
plt.imshow(cv.cvtColor(output, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
