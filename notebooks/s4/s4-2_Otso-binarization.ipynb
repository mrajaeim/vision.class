```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
car_num_img = cv.imread("../../images/car_number.jpg", 0)

plt.figure(figsize=(20, 8))
plt.subplot(241); plt.imshow(car_num_img, cmap="gray", vmin=0, vmax=255); plt.title("Original")
plt.subplot(242); plt.hist(car_num_img.ravel(), bins=256, range=[0, 256]); plt.title("Original Histogram")

threshold = 127
ret, res = cv.threshold(car_num_img, threshold, 255, cv.THRESH_BINARY) 

ret, res = cv.threshold(car_num_img, threshold, 255, cv.THRESH_BINARY_INV+cv.THRESH_OTSU)

blured_img = cv.GaussianBlur(car_num_img, (5,5), 0)
ret, res = cv.threshold(blured_img, threshold, 255, cv.THRESH_BINARY_INV+cv.THRESH_OTSU) 


plt.subplot(243); plt.imshow(res, cmap="gray", vmin=0, vmax=255); plt.title("BINARY")
plt.subplot(244); plt.hist(res.ravel(), bins=256, range=[0, 256]); plt.title("BINARY Histogram")

plt.subplot(245); plt.imshow(res, cmap="gray", vmin=0, vmax=255); plt.title("Otso")
plt.subplot(246); plt.hist(res.ravel(), bins=256, range=[0, 256]); plt.title("Otso Histogram")

plt.subplot(247); plt.imshow(res, cmap="gray", vmin=0, vmax=255); plt.title("Blured Otso")
plt.subplot(248); plt.hist(res.ravel(), bins=256, range=[0, 256]); plt.title("Blured Otso Histogram")
```
