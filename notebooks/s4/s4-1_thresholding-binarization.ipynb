```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
```

```python
img = cv.imread("../../images/gradient.png", 0)

print(img.shape)

plt.figure(figsize=(20, 8))
plt.subplot(131); plt.imshow(img, cmap="gray"); plt.title("Original")

binary_img = img.copy()
threshold = 127
binary_img[binary_img < threshold] = 0
binary_img[binary_img >= threshold] = 255
plt.subplot(132); plt.imshow(binary_img, cmap="gray"); plt.title("Binary")

inv_binary_img = cv.bitwise_not(binary_img)
plt.subplot(133); plt.imshow(inv_binary_img, cmap="gray"); plt.title("Inverted Binary")
```

```python
img = cv.imread("../../images/gradient.png", 0)

print(img.shape)

plt.figure(figsize=(20, 8))
plt.subplot(131); plt.imshow(img, cmap="gray"); plt.title("Original")

threshold = 127
ret, binary_img = cv.threshold(img, threshold, 255, cv.THRESH_BINARY) 
plt.subplot(132); plt.imshow(binary_img, cmap="gray"); plt.title("Binary")

ret, inv_binary_img = cv.threshold(img, threshold, 255, cv.THRESH_BINARY_INV)
plt.subplot(133); plt.imshow(inv_binary_img, cmap="gray"); plt.title("Inverted Binary")
```

```python
img = cv.imread("../../images/gradient.png", 0)

print(img.shape)

plt.figure(figsize=(20, 8))
plt.subplot(231); plt.imshow(img, cmap="gray", vmin=0, vmax=255); plt.title("Original")

threshold = 127
ret, binary_img = cv.threshold(img, threshold, 255, cv.THRESH_BINARY) 
plt.subplot(232); plt.imshow(binary_img, cmap="gray", vmin=0, vmax=255); plt.title("BINARY")

ret, inv_binary_img = cv.threshold(img, threshold, 255, cv.THRESH_BINARY_INV)
plt.subplot(233); plt.imshow(inv_binary_img, cmap="gray", vmin=0, vmax=255); plt.title("BINARY_INV")

ret, binary_img = cv.threshold(img, threshold, 255, cv.THRESH_TRUNC) 
plt.subplot(234); plt.imshow(binary_img, cmap="gray", vmin=0, vmax=255); plt.title("TRUNC")

ret, inv_binary_img = cv.threshold(img, threshold, 255, cv.THRESH_TOZERO)
plt.subplot(235); plt.imshow(inv_binary_img, cmap="gray", vmin=0, vmax=255); plt.title("TOZERO")

ret, binary_img = cv.threshold(img, threshold, 255, cv.THRESH_TOZERO_INV) 
plt.subplot(236); plt.imshow(binary_img, cmap="gray", vmin=0, vmax=255); plt.title("TOZERO_INV")
```

```python
import cv2
import numpy as np

# Create a simple test image
image = np.array([[50, 150, 200],
                  [100, 127, 255],
                  [30, 180, 220]], dtype=np.uint8)

threshold_value = 127
max_value = 255

ret, result = cv2.threshold(image, threshold_value, max_value, cv2.THRESH_TRUNC)

print("Original Image:")
print(image)
print("\nAfter THRESH_TRUNC (threshold=127):")
print(result)

plt.figure(figsize=(20, 8))
plt.subplot(121); plt.imshow(image, cmap="gray", vmin=0, vmax=255); plt.title("Original"); plt.axis('image')
plt.subplot(122); plt.imshow(result, cmap="gray", vmin=0, vmax=255); plt.title("Truncate"); plt.axis('image')

print(image.shape, image.dtype)
print(result.shape, result.dtype)

```
