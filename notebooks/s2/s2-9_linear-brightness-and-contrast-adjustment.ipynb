```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../../images/2.png")

alpha_beta = [
    [2, 20],
    [3, -150],
    [2, -30]
]

plt.figure(figsize=(20,7))
plt.subplot(241); plt.imshow(img[...,::-1]); plt.title("Original")
for idx, (a, b) in enumerate(alpha_beta):
    plt.subplot(242+idx)
    plt.imshow(
        cv2.convertScaleAbs(img, alpha=a, beta=b)[...,::-1]
    )
    plt.title(f"alpha={a}, beta={b}")

low_contrast_img = cv2.convertScaleAbs(img, alpha=0.5, beta=50)
high_contrast_img = cv2.convertScaleAbs(img, alpha=2, beta=-50)
plt.subplot(246); plt.imshow(high_contrast_img); plt.title("high_contrast_img")
plt.subplot(247); plt.imshow(low_contrast_img); plt.title("low_contrast_img")
```

## Histogram analysis

```python
gray_scale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_scale_high_contrast_img = cv2.cvtColor(high_contrast_img, cv2.COLOR_BGR2GRAY)
gray_scale_low_contrast_img = cv2.cvtColor(low_contrast_img, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(20,5))
plt.subplot(131); plt.hist(gray_scale_img.ravel(), 256, [0, 255]); plt.title("img")
plt.subplot(132); plt.hist(gray_scale_low_contrast_img.ravel(), 256, [0, 255]); plt.title("low_contrast_img")
plt.subplot(133); plt.hist(gray_scale_high_contrast_img.ravel(), 256, [0, 255]); plt.title("high_contrast_img")
```
