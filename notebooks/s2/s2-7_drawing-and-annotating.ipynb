```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_size = (512, 512, 3)
img = np.zeros(img_size, np.uint8)
img_bw = np.zeros(img_size[:2], np.uint8)

fig, axes = plt.subplots(1,2)
axes[0].imshow(img[...,::-1]) ; axes[0].set_title("3 Channel")
axes[1].imshow(img_bw, cmap="gray"); axes[1].set_title("1 Channel")
```

```python
img = np.zeros(img_size, np.uint8)
cv2.line(img, (0,0), img_size[:2], (255, 127, 0), 5)
plt.imshow(img[...,::-1])
```

```python
img = np.zeros(img_size, np.uint8)
cv2.rectangle(img, (100,100), (200, 200), (255, 127, 0), 5)
cv2.rectangle(img, (200,200), (300, 300), (0, 127, 255), -1)
plt.imshow(img[...,::-1])
```

```python
img = np.zeros(img_size, np.uint8)
cv2.circle(img, (100,100), 100, (255, 127, 0), 5)
cv2.circle(img, (300,300), 150, (0, 127, 255), -1)
plt.imshow(img[...,::-1])
```

```python
img = np.zeros(img_size, np.uint8)
cv2.ellipse(img, (100, 100), (100, 150), 30, 0, 360, (255, 127, 0), 5)
cv2.ellipse(img, (250, 250), (150, 100), 30, 0, 180, (0, 127, 255), -1)
plt.imshow(img[...,::-1])
```

```python
img = np.zeros(img_size, np.uint8)
lines = [
    np.array([[30, 40], [55, 80], [100, 250], [300, 400], [0, 300], [0, 0]], np.int32)
]
for i in range(1, 10):
    lines.append(
        lines[i-1] + 15
    )
cv2.polylines(img, lines, True, (255, 127, 0), 5)
plt.imshow(img[...,::-1])
```

## Text on image

```python
img = np.zeros(img_size, np.uint8)
for i in range(10):
    cv2.putText(img, "Hello world!", (50, 75+(i*45)), cv2.FONT_HERSHEY_COMPLEX, 2, (0, 0, 255), 1+i)
plt.imshow(img[...,::-1])
```

```python
img = np.zeros(img_size, np.uint8)
text = " "
cv2.putText(img, text, (50, 150), cv2.FONT_HERSHEY_COMPLEX, 2, (0, 0, 255), 3)
plt.imshow(img[...,::-1])
```
