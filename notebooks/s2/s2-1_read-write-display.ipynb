```python
import cv2
import numpy as np
```

```python
img = cv2.imread("../../images/1.png")

cv2.imshow("It's a image", img)

wait_time = 5000 #ms
key = cv2.waitKey(wait_time)
print(key)

cv2.destroyAllWindows()
```

```python
print(type(img))
print(img.shape)
```

Image data is typically formatted as (H, W, C) – height, width, and channels – but operating systems often represent image shapes as (W, H). This difference arises because NumPy, a common library for numerical computation, uses matrices where rows represent height and columns represent width. Therefore, the number of rows corresponds to the image's height, and the number of columns corresponds to its width

```python
print(f"Height (rows) of image: {img.shape[0]}px")
print(f"Width (columns) of image: {img.shape[1]}px")
```

```python
cv2.imwrite("../../output.jpg", img)
result = cv2.imwrite("../../output.png", img)

print(result)
```

A `true` result indicates successful image saving; `false` suggests potential errors like insufficient permissions, disk space, or an incorrect file path.

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18,4))
axes[0].imshow(img[:,:,::-1]); axes[0].set_title("M1: [:, :, ::-1]") 
axes[1].imshow(img[...,::-1]); axes[1].set_title("M2: [..., ::-1]")
axes[2].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); axes[2].set_title("M3: cv2cvtColor(img, cv2.COLOR_BGR2RGB)")
fig.suptitle("Rearrange to RGB")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,4))
axes[0].imshow(img[:,::-1,::-1]); axes[0].set_title("Horizontal flip")
axes[1].imshow(img[::-1,:,::-1]); axes[1].set_title("Vertical flip")
```
