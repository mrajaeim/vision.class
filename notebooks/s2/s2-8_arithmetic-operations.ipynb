```python
a = np.ones([2], dtype=np.uint8)
print(a)
b = a * 100
print(b)
print(b + 100)
print(b + 160)
```

```python
m1 = np.array([[150, 156], [167, 169]], dtype=np.uint8)
m2 = np.ones([2,2], dtype=np.uint8) * 100

m1 = m1.astype(np.int16) + m2
m1 = np.minimum(m1, 255)
m1 = m1.astype(np.uint8)

print(m1, m1.dtype)
```

```python
m1 = np.array([[150, 156], [167, 169]], dtype=np.uint8)
value_to_add = 100

max_threshold = 255 - value_to_add

m1[m1 >= max_threshold] = 255
m1[m1 < max_threshold] += value_to_add

print(m1, m1.dtype)
```

## Use open-cv arithmetic mehtods

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("../../images/2.png")

M = np.full(img.shape, 100, dtype=np.uint8) 

added = cv2.add(img, M)
subtracted = cv2.subtract(img, M)

plt.figure(figsize=(20,5))

plt.subplot(131);plt.imshow(img[...,::-1]);plt.title("Original")
plt.subplot(132);plt.imshow(added[...,::-1]);plt.title("Added")
plt.subplot(133);plt.imshow(subtracted[...,::-1]);plt.title("Subtracted")
```

```python
img = cv2.imread("../../images/1.png")
logo_img = cv2.imread("../../images/logo.png")

modified_img = img[400: 400+logo_img.shape[0], 800: 800+logo_img.shape[1], :]
modified_img = cv2.add(modified_img, logo_img)
plt.imshow(modified_img[...,::-1])
```

```python
img = cv2.imread("../../images/1.png")
logo_img = cv2.imread("../../images/logo.png")

output_img = np.zeros_like(img, dtype=np.uint8)
output_img[400: 400+logo_img.shape[0], 800: 800+logo_img.shape[1], :] = logo_img
plt.imshow(output_img[...,::-1])
```

```python
output_img = cv2.add(img, output_img)
plt.imshow(output_img[...,::-1])
```

```python
img1 = cv2.imread("../../images/logo.png")
img2 = cv2.imread("../../images/logo-2.png")

output_image = cv2.addWeighted(img1, .4, img2, .6, 0)
plt.imshow(output_image[...,::-1])
```
